```python
import math

def sinusoidal_behavior_model(time, A0, B0, r, R, R_prime, C, age, experience, risk_management):

    # Calcul des facteurs externes
    A_t=A0 * math.exp(-time) * math.exp(-r/R)
    B_t=B0 * math.exp(-time) * math.exp(-r/R_prime)
    external_factor = A_t + B_t  + C

    # Oscillations sinuso√Ødales pour le stress et la perception du danger
    stress_level = (external_factor/(age + risk_management + experience)) * math.sin(time)
    perception_of_danger = (external_factor/(age + risk_management + experience)) * math.cos(time)

    return stress_level, perception_of_danger

# Exemple d'utilisation
stress, danger_perception = sinusoidal_behavior_model(time=1, A0=10, B0=20, r=5, R=100, R_prime=80, C=50, age=30, experience= 40, risk_management= 50)
print(stress, danger_perception)
```

```python
import math
import matplotlib.pyplot as plt

# Constants
A_0 = 10
B_0 = 20
r = 5
R = 100
R_prime = 80
C = 50
experience = 5
risk_management = 8
time = 1

# Simulating for different ages
ages = range(18, 66)  # Age from 18 to 65
stress_levels = []
danger_perceptions = []

for age in ages:
    stress, danger_perception = sinusoidal_behavior_model(time, A_0, B_0, r, R, R_prime, C, age, experience, risk_management)
    stress_levels.append(stress)
    danger_perceptions.append(danger_perception)

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(ages, stress_levels, label='Stress Level')
plt.xlabel('Experience')
plt.ylabel('Stress Level')
plt.title('Stress Level vs experience')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(ages, danger_perceptions, label='Perception of Danger')
plt.xlabel('Experience')
plt.ylabel('Perception of Danger')
plt.title('Perception of Danger vs experience')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
